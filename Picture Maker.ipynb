```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
IMG_DIR = "/Users/brentredmon/Downloads/5layers"
```

```python
loss = Image.open(IMG_DIR + "/loss.png")
mrcnn_bbox_loss = Image.open(IMG_DIR + "/mrcnn_bbox_loss.png")
mrcnn_class_loss = Image.open(IMG_DIR + "/mrcnn_class_loss.png")
mrcnn_mask_loss = Image.open(IMG_DIR + "/mrcnn_mask_loss.png")
rpn_bbox_loss = Image.open(IMG_DIR + "/rpn_bbox_loss.png")
rpn_class_loss = Image.open(IMG_DIR + "/rpn_class_loss.png")
val_loss = Image.open(IMG_DIR + "/val_loss.png")
val_mrcnn_bbox_loss = Image.open(IMG_DIR + "/val_mrcnn_bbox_loss.png")
val_mrcnn_class_loss = Image.open(IMG_DIR + "/val_mrcnn_class_loss.png")
val_mrcnn_mask_loss = Image.open(IMG_DIR + "/val_mrcnn_mask_loss.png")
val_rpn_bbox_loss = Image.open(IMG_DIR + "/val_rpn_bbox_loss.png")
val_rpn_class_loss = Image.open(IMG_DIR + "/val_rpn_class_loss.png")
```

```python
plt.figure(figsize=(18, 16))

plt.subplot(1, 3, 1)
plt.title("Loss")
plt.axis('off')
plt.imshow(loss)

plt.subplot(1, 3, 2)
plt.title("MRCNN Bbox loss")
plt.axis('off')
plt.imshow(mrcnn_bbox_loss)

plt.subplot(1, 3, 3)
plt.title("MRCNN Class loss")
plt.axis('off')
plt.imshow(mrcnn_class_loss)
```

```python
fig,axes= plt.subplots(nrows=3, ncols= 4, figsize=(18, 16))

axes[0][0].imshow(loss)
axes[0][0].set_title('loss')
axes[0][0].axis('off')

axes[0][1].imshow(mrcnn_bbox_loss)
axes[0][1].set_title('mrcnn_bbox_loss')
axes[0][1].axis('off')

axes[0][2].imshow(mrcnn_class_loss)
axes[0][2].set_title('mrcnn_class_loss')
axes[0][2].axis('off')

axes[0][3].imshow(mrcnn_mask_loss)
axes[0][3].set_title('mrcnn_mask_loss')
axes[0][3].axis('off')


axes[1][0].imshow(rpn_bbox_loss)
axes[1][0].set_title('rpn_bbox_loss')
axes[1][0].axis('off')

axes[1][1].imshow(rpn_class_loss)
axes[1][1].set_title('rpn_class_loss')
axes[1][1].axis('off')

axes[1][2].imshow(val_loss)
axes[1][2].set_title('val_loss')
axes[1][2].axis('off')

axes[1][3].imshow(val_mrcnn_bbox_loss)
axes[1][3].set_title('val_mrcnn_bbox_loss')
axes[1][3].axis('off')


axes[2][0].imshow(val_mrcnn_class_loss)
axes[2][0].set_title('val_mrcnn_class_loss')
axes[2][0].axis('off')

axes[2][1].imshow(val_mrcnn_mask_loss)
axes[2][1].set_title('val_mrcnn_mask_loss')
axes[2][1].axis('off')

axes[2][2].imshow(val_rpn_bbox_loss)
axes[2][2].set_title('val_rpn_bbox_loss')
axes[2][2].axis('off')

axes[2][3].imshow(val_rpn_class_loss)
axes[2][3].set_title('val_rpn_class_loss')
axes[2][3].axis('off')


plt.show()
```
